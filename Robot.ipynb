```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = np.array([6, 8, 10, 14, 18]) #diametro em polegadas das pizzas
y = np.array([7, 9, 13, 17.5, 18]) #pre√ßo em reais
```

```python
plt.scatter(x, y) #criar gr√°fico
plt.xlabel('Di√¢metro em polegadas')
plt.ylabel('Pre√ßo em reais')
plt.title('Pre√ßo Pizza em raz√£o ao seu di√¢metro')
plt.show()
```

```python
model = LinearRegression()
x = x.reshape(-1, 1) #Transformar o array x em uma matriz de coluna
model.fit(x, y) #Treinar o modelo com os dados
```

```python
y_pred = model.predict(x) #fazer as previs√µes para os dados de treino
mse = mean_squared_error(y,y_pred) #calcular o erro quadr√°tico m√©dio
r2 = r2_score(y, y_pred) #calcular o coeficiente de determina√ß√£o
print(f'MSE= {mse:.2f}') #imprimir MSE com duas casas decimais
print(f'R2= {r2:.2f}') #imprimir R2 com duas casas decimais
```

```python
x_new = np.array([12]) #Mudar de acordo com o diametro da pizza
y_new = model.predict(x_new.reshape(-1,1)) #fazer previs√£o para um nova pizza
```

```python
print(f'Uma pizza de {x_new[0]} polegadas custa ${y_new[0]: .2f}')
```
